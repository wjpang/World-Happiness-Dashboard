```python
whr19_20 = pandas.read_csv("data_clean/WHR_data_clean.csv")
```

```python
whr19_20 = whr19_20[whr19_20.Year >= 2019]
whr19_20
```

```python
whr19_20.drop("log(GDP per Capita)", inplace=True, axis=1)
whr19_20.drop("Social Support", inplace=True, axis=1)
whr19_20.drop("Life Expectancy", inplace=True, axis=1)
whr19_20.drop("Freedom Score", inplace=True, axis=1)
whr19_20.drop("Generosity", inplace=True, axis=1)
whr19_20.drop("Perceptions of Corruption", inplace=True, axis=1)
whr19_20.reset_index(inplace=True)
whr19_20.drop("index", inplace=True, axis=1)
whr19_20
```

```python
whr19_20 = whr19_20[whr19_20.duplicated(subset=["Country"], keep=False)]
whr19_20 = whr19_20.reset_index()
whr19_20 = whr19_20.drop("index", axis=1)
whr19_20
```

```python
whr19_20 = whr19_20.pivot_table(index="Country", columns="Year", values="Happiness Score").reset_index()
whr19_20
```

```python
whr19_20.rename(columns={2019: "Year_2019", 2020: "Year_2020"}, inplace=True)
whr19_20
```

```python
whr19_20["Change in Happiness"] = whr19_20["Year_2020"] - whr19_20["Year_2019"]
whr19_20
```

```python
whr19_20.drop("Year_2019", inplace=True, axis=1)
whr19_20.drop("Year_2020", inplace=True, axis=1)
whr19_20.reset_index(inplace=True)
whr19_20.drop("index", inplace=True, axis=1)
whr19_20
```

```python
country_list = whr19_20["Country"]
mortality = mortality[mortality["Country"].isin(country_list)]
mortality = mortality.reset_index()
mortality.drop("index", inplace=True, axis=1)
mortality
```

```python
mortality.drop("Island", inplace=True, axis=1)
mortality.drop("Median age", inplace=True, axis=1)
mortality.drop("log(Average Distance to SARS countries)", inplace=True, axis=1)
mortality.drop("WHO Western Pacific Region", inplace=True, axis=1)
mortality.drop("Female head of government", inplace=True, axis=1)
mortality.drop("Index of institutional trust", inplace=True, axis=1)
mortality.drop("Gini coefficient of income", inplace=True, axis=1)
mortality.drop("All-cause death count, 2017", inplace=True, axis=1)
mortality.drop("All-cause death count, 2018", inplace=True, axis=1)
mortality.drop("All-cause death count, 2019", inplace=True, axis=1)
mortality.drop("All-cause death count, 2020", inplace=True, axis=1)
mortality
```

```python
whrcovid = pandas.merge(whr19_20, mortality, how="outer", on="Country")
whrcovid
```

```python
whrcovid.drop("Population 2019", inplace=True, axis=1)
whrcovid
```

```python
whrcovid.replace("Hong Kong S.A.R. of China", "Hong Kong", inplace=True)
whrcovid.replace("Taiwan Province of China", "Taiwan", inplace=True)
```

```python
import pycountry_convert as pc

def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

continent_lst = []
for country in whrcovid["Country"]:
    if country == "Kosovo":
        continent_lst.append("Europe")
    else:
        continent_lst.append(country_to_continent(country))
print(continent_lst)
```

```python
whrcovid["Continent"] = continent_lst
whrcovid
```

```python
whrcovid.to_csv("data_clean/WHR_covid_clean.csv", index=False)
```
