```python
import pandas
```

```python
whr = pandas.read_excel("data_raw/WHR2021-DataPanel.xls")
whr
```

```python
whr.drop("Positive affect", inplace=True, axis=1)
whr.drop("Negative affect", inplace=True, axis=1)
whr
```

```python
whr.rename(columns={"Country name": "Country", "year": "Year", "Life Ladder": "Happiness Score", "Healthy life expectancy at birth": "Life Expectancy", "Social support": "Social Support", "Freedom to make life choices": "Freedom Score", "Perceptions of corruption": "Perceptions of Corruption", "Log GDP per capita": "log(GDP per Capita)"}, inplace=True)
whr
```

```python
whr.to_csv("data_clean/WHR_data_clean.csv", index=False)
```

```python
mortality = pandas.read_excel("data_raw/WHR2021-MortalityData.xlsx")
mortality
```

```python
mortality["Island"] = mortality["Island"].astype(bool)
mortality["WHO Western Pacific Region"] = mortality["WHO Western Pacific Region"].astype(bool)
mortality["Female head of government"] = mortality["Female head of government"].astype(bool)
mortality
```

```python
mortality.rename(columns={"Country name": "Country", "Index of exposure to COVID-19  infections in other countries as of March 31": "Index of exposure to COVID-19 infections in other countries", "Log of average distance to SARS countries": "log(Average Distance to SARS countries)"}, inplace=True)
mortality
```

```python
mortality.to_csv("data_clean/WHR_mortality_clean.csv", index=False)
```
