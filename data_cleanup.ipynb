```python
whr19_20 = pandas.read_csv("data_clean/WHR_data_clean.csv")
```

```python
whr19_20 = whr19_20[whr19_20.Year >= 2019]
whr19_20
```

```python
whr19_20.drop("log(GDP per Capita)", inplace=True, axis=1)
whr19_20.drop("Social Support", inplace=True, axis=1)
whr19_20.drop("Life Expectancy", inplace=True, axis=1)
whr19_20.drop("Freedom Score", inplace=True, axis=1)
whr19_20.drop("Generosity", inplace=True, axis=1)
whr19_20.drop("Perceptions of Corruption", inplace=True, axis=1)
whr19_20.reset_index(inplace=True)
whr19_20.drop("index", inplace=True, axis=1)
whr19_20
```

```python
whr19_20 = whr19_20[whr19_20.duplicated(subset=["Country"], keep=False)]
whr19_20 = whr19_20.reset_index()
whr19_20 = whr19_20.drop("index", axis=1)
whr19_20
```

```python
whr19_20 = whr19_20.pivot_table(index="Country", columns="Year", values="Happiness Score").reset_index()
whr19_20
```
